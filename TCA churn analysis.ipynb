```python
import pandas as pd

# Load dataframe (make sure this was already done)
df = pd.read_csv("/content/telcocustomerdata.csv")   # or however you loaded it earlier

def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)

df.head(30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize =(3,4))
ax = sns.countplot(x='Churn',data = df)
ax.bar_label(ax.containers[0])
plt.title("Churn Distribution of customer ")
plt.show()
```

```python
plt.figure(figsize =(3,4))
gb = df["Churn"].value_counts()   # Get counts of Yes/No
plt.pie(gb, labels=gb.index, autopct='%1.1f%%')
plt.title("Churn Distribution of customer in percentage")
plt.title("Churn Distribution")
plt.show()
```

```python
# from given pie we can calculate 26.5 customer churned out
# now explored reason behind it
```

```python
import matplotlib.pyplot as plt

# Pivot table to get counts
pivot = df.groupby('SeniorCitizen')['Churn'].value_counts().unstack().fillna(0)

# Convert to percentage
percent_data = pivot.div(pivot.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = percent_data.plot(kind='bar', stacked=True, figsize=(8, 6))

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

plt.title("Churn Percentage Distribution by SeniorCitizen")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", loc="upper right")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of service-related columns
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Number of columns in grid (change if needed)
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate rows dynamically

# Create figure and subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
axes = axes.flatten()  # Flatten for easy looping

# Loop through columns and plot
for i, col in enumerate(columns):
    sns.countplot(x=col, hue='Churn', data=df, ax=axes[i])
    axes[i].set_title(f"Count Plot of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")

# Remove extra empty subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
#ustomers who do not have services like OnlineSecurity, TechSupport, OnlineBackup, and DeviceProtection show significantly higher churn. Additionally, fiber optic internet users churn more compared to DSL users, indicating dissatisfaction or competitive alternatives. No-internet-service users show very low churn, and MultipleLines provides slight retention but is not a strong churn influencer. Overall, security and support-related services strongly correlate with customer retention.
```
